```python
import pandas as pd
```

```python
df = pd.read_csv('/Crop_recommendation.csv')
df.head(6)
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts()
```

```python
x = df.drop('label', axis= 1)
y = df['label']
```

```python
x.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.2)
```

```python
x_train.info()
```

```python
y_train.info()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
model1.fit(x_train, y_train)
```

```python
y_pred1 = model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

logistic_regression_accuracy = accuracy_score(y_test, y_pred1)

print("logistic_regression_accuracy: " + str(logistic_regression_accuracy))
```

```python
from sklearn.tree import DecisionTreeClassifier

model2 = DecisionTreeClassifier()

model2.fit(x_train, y_train)

y_pred2 = model2.predict(x_test)
```

```python
decision_tree_classifier_accuracy = accuracy_score(y_test, y_pred2)

print("decision_tree_classifier_accuracy: " + str(decision_tree_classifier_accuracy))
```
